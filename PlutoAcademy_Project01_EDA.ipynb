````python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")
print("Dataset Loaded Succesfully!")
print("Rows and Columns:",df.shape)
df.head()
````

````python
df.info()
````

````python
df.isnull().sum()
````

````python
print("Duplicates before:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("Duplicates after:", df.duplicated().sum())
````

The Netflix dataset contains 8,807 records and 12 columns. It includes information about movies and TV shows available on Netflix. The dataset contains details such as title, director, cast, country, release year, rating, and genre. Missing values were found in a few columns and were handled during data cleaning. No duplicate records were found in the dataset, making it suitable for exploratory data analysis and visualization.

````python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

df.isnull().sum()
````

````python
df['date_added'] = df['date_added'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not Rated')
df['duration'] = df['duration'].fillna('Unknown')

df.isnull().sum()
````

### Data Cleaning Decisions

During data cleaning, I found missing values in the director, cast, country, date_added, rating, and duration columns. To avoid losing useful information, I replaced the missing values with appropriate labels such as "Unknown" and "Not Rated" instead of deleting the records. This allowed me to keep all 8,807 entries in the dataset for analysis. I also checked the dataset for duplicate records using the duplicated() function and found that there were no duplicate rows, so no records were removed. After applying these changes, I verified the dataset again and confirmed that all missing values had been handled successfully, making the data clean and ready for exploratory data analysis and visualization.
